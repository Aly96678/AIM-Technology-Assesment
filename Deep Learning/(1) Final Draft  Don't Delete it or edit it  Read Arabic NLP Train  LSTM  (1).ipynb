```python
from collections import Counter
```

```python


def counter_word(text):
    count = Counter()
    for i in text.values:
        for word in i.split():
            count[word] +=1
    return count            
```

```python
text = X_train
counter = counter_word(text)
```

```python
num_word = len(counter)

max_length = 20
```

# Tokinez the data set 

```python
from keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer(num_words=num_word)
tokenizer.fit_on_texts(X_train)
```

```python
word_index = tokenizer.word_index
```

```python
word_index
```

```python
num_words = len(word_index)
```

```python
num_words
```

```python
train_sequences = tokenizer.texts_to_sequences(X_train)
```

```python
train_sequences[0]
```

## Train Model

```python
from keras.preprocessing.sequence import pad_sequences
```

```python
max_length = 20
```

```python
train_padded = pad_sequences(train_sequences, maxlen=max_length,padding='post',truncating='post')
```

```python
train_padded[0]
```

```python
test_sequences = tokenizer.texts_to_sequences(X_test)
test_padded = pad_sequences(test_sequences, maxlen=max_length,padding='post',truncating='post')
```

```python
test_padded
```

```python
print(X_train[0])
```

```python
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense, Dropout
from keras.initializers import Constant
from tensorflow.keras.optimizers import Adam
```

```python


model = Sequential()

model.add(Embedding(num_words,32,input_length=max_length))
model.add(LSTM(64,dropout=0.1))
model.add(Dense(18,activation='sigmoid'))

optimizer = Adam(learning_rate=3e-4)


model.compile(loss='binary_crossentropy',optimizer=optimizer, metrics=['accuracy'])
```

```python
# Precision 
from sklearn.metrics import precision_score
precision = precision_score(y_test_number, ynew, average='micro')
print('Precision: %.3f' % precision)
```

```python
# Recall
from sklearn.metrics import recall_score
recall = recall_score(y_test_number, ynew, average='micro')
print('Recall: %.3f' % recall)
```

```python
from sklearn.metrics import f1_score

score = f1_score(y_test_number, ynew, average='weighted')
print('F1_Score: %.3f' % score)
```
