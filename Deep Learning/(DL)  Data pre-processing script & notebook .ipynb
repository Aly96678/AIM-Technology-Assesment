# Remove all Characters except Arabic Character 

```python
def remove_english_string(string):
    return " ".join(re.findall('[\u0600-\u06ff]+', string))
  
```

```python
from collections import Counter
```

```python
# Count number of the word in the dataset
def counter_word(text):
    count = Counter()
    for i in text.values:
        for word in i.split():
            count[word] +=1
    return count    
```
